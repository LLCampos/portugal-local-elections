```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import pearsonr


%matplotlib inline

plt.rc('axes', labelsize=20) 
plt.rcParams["figure.figsize"] = (10,10)
plt.rc('xtick', labelsize=13)
plt.rc('ytick', labelsize=13) 
mpl.rcParams['axes.titlesize'] = 24
```

```python
xls_file = pd.ExcelFile('data/raw_data/AbstencaoPresidenciais2016.xlsx')
eleicoes_abstencao = xls_file.parse('Sheet1')
```

```python
# Create %Abstention Column
eleicoes_abstencao['Abstention (%)'] = 100 - ((eleicoes_abstencao['Abstencao'] / eleicoes_abstencao.Inscritos) * 100)
```

```python
# Correlation using all the data
plot = plt.scatter(eleicoes_abstencao.Inscritos, eleicoes_abstencao['Abstention (%)'])
plt.xlabel('Number of Enrolled Voters')
plt.ylabel('Abstention (%)')
plt.title('Presidential Elections 2015')

# Pearson Correlation Coefficient and p-value
corr_cof, p_value = pearsonr(eleicoes_abstencao.Inscritos, eleicoes_abstencao['Abstention (%)'])
print 'Correlation Cofficient: {}'.format(corr_cof)
print 'p-value: {}'.format(p_value)
```

```python
# What if I used the log of the Number of Enrolled Voters?
plot = plt.scatter(np.log(eleicoes_abstencao.Inscritos), eleicoes_abstencao['Abstention (%)'])
plt.xlabel('Natural Log Number of Enrolled Voters')
plt.ylabel('Abstention (%)')
plt.title('Presidential Elections 2015')

# Pearson Correlation Coefficient and p-value
corr_cof, p_value = pearsonr(np.log(eleicoes_abstencao.Inscritos), eleicoes_abstencao['Abstention (%)'])
print 'Correlation Cofficient: {}'.format(corr_cof)
print 'p-value: {}'.format(p_value)
```
